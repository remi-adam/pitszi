```python
import sys
sys.path.insert(0,'/Users/adam/Project/NIKA/Software/Processing/Labtools/RA/pitszi/')
from astropy.io import fits
from astropy import units as u
from astropy.wcs import WCS
from astropy.coordinates import SkyCoord
from scipy.optimize import curve_fit
from scipy.ndimage import gaussian_filter
import numpy as np
import matplotlib.pyplot as plt

import pitszi
```

# Input data

```python
dir_sim  = '/Users/adam/Project/Notes-Papier-Conf/2016_12_Edge_Search/Save/HydroSimu/Selected/'

'''
clname = 'RG474_00235_Ymap_zobs0p4'
redshift = 0.39
M500 = 12.7*1e14*u.Msun
'''

clname = 'RG377_00181_Ymap_zobs0p5'
redshift = 0.54
M500 = 3.7*1e14*u.Msun

beamFWHM = 6.0
```

```python
hdul = fits.open(dir_sim+clname+'_regrid.fits')
dat_img = hdul[0].data/(-12.0)
dat_head = hdul[0].header

dat_img = gaussian_filter(dat_img, sigma=beamFWHM/2.35/(dat_head['CDELT2']*3600))
```

# Define the model

```python
cluster = pitszi.Model(RA=dat_head['CRVAL1']*u.deg, Dec=dat_head['CRVAL2']*u.deg, 
                       redshift=redshift, M500=M500, 
                       silent=False, output_dir='/Users/adam/Desktop/pitszi_output')

cluster.set_pressure_profile_universal_param('A10MD')

cluster.model_pressure_fluctuation = {'name': 'CutoffPowerLaw',
                                      'Norm': 0.5, 'slope': -11/3.0,
                                      'Linj': 1*u.Mpc,'Ldis': 1*u.kpc}

cluster.triaxiality = {'min_to_maj_axis_ratio':1, 'int_to_maj_axis_ratio':1,
                       'euler_angle1':0*u.deg, 'euler_angle2':0*u.deg, 'euler_angle3':0*u.deg}
        
cluster.map_header =  dat_head
cluster.los_reso  = 20*u.kpc
cluster.los_size  = 5*u.Mpc
cluster.R_truncation = cluster.R500*5
```

```python
mock_image = cluster.get_sz_map()
```

```python
plt.rcParams.update({'font.size': 10})
fig = plt.figure(0, figsize=(18, 8))
ax = plt.subplot(1, 2, 1, projection=WCS(dat_head))
plt.imshow(mock_image*1e5)
plt.colorbar()

ax = plt.subplot(1, 2, 2, projection=WCS(cluster.get_map_header()))
plt.imshow(dat_img*1e5)
plt.colorbar()
```

# Fit for the mean pressure profile and shape

```python
#===== Fitting function
def fitfunc(xdata, 
            par1, par2, par3,
            par4, par5, 
            par6, par7):
    cluster.silent=True
    cluster.model_pressure_profile = {'name': 'beta',
                                      'n_0': par1*u.keV*u.cm**-3,
                                      'r_c': par2*u.kpc,
                                      'beta': par3}
    # Only assume a ellipticity along the plane of the sky
    cluster.triaxiality = {'min_to_maj_axis_ratio':par4, 'int_to_maj_axis_ratio':par4,
                       'euler_angle1':0*u.deg, 'euler_angle2':90*u.deg, 'euler_angle3':par5*u.deg}
    
    cluster.coord = SkyCoord((dat_head['CRVAL1']+par6)*u.deg, 
                             (dat_head['CRVAL2']+par7)*u.deg, frame='icrs')
    
    model = cluster.get_sz_map(no_fluctuations=True)
    return np.reshape(model, model.size)
```

```python
#===== SZ fit for the processed data
xdata = np.zeros(1)
p0 = [0.01, 500, 0.5, 1, 0, 0, 0]

bounds = ([0,      0,      -np.inf, 0, -90, -np.inf, -np.inf],
          [np.inf, np.inf, +np.inf, 1, +90, +np.inf, +np.inf])
popt_s, pcov = curve_fit(fitfunc, xdata, np.reshape(dat_img, dat_img.size), p0=p0, 
                       bounds=bounds)
print(popt_s)
```

```python
#===== Set the cluster to best fit
cluster.model_pressure_profile = {'name': 'beta',
                                  'n_0': popt_s[0]*u.keV*u.cm**-3,
                                  'r_c': popt_s[1]*u.kpc,
                                  'beta': popt_s[2]}

cluster.triaxiality = {'min_to_maj_axis_ratio':popt_s[3], 'int_to_maj_axis_ratio':popt_s[3],
                       'euler_angle1':0*u.deg, 'euler_angle2':90*u.deg, 'euler_angle3':popt_s[4]*u.deg}

cluster.coord = SkyCoord((dat_head['CRVAL1']+popt_s[5])*u.deg, 
                         (dat_head['CRVAL2']+popt_s[6])*u.deg, frame='icrs')

best_model = cluster.get_sz_map(no_fluctuations=True)
```

```python
#===== Show the residual
plt.rcParams.update({'font.size': 10})
fig = plt.figure(0, figsize=(15, 15))
ax = plt.subplot(2, 2, 1, projection=WCS(dat_head))
plt.imshow(dat_img*1e5)
plt.colorbar()

ax = plt.subplot(2, 2, 2, projection=WCS(cluster.get_map_header()))
plt.imshow(best_model*1e5)
plt.colorbar()

ax = plt.subplot(2, 2, 3, projection=WCS(cluster.get_map_header()))
plt.imshow((dat_img-best_model)*1e5)
plt.colorbar()

ax = plt.subplot(2, 2, 4, projection=WCS(cluster.get_map_header()))
plt.imshow((dat_img-best_model)/best_model, vmin=-1, vmax=1)
plt.colorbar()
```

# Extract the power spectrum of the fluctuations

```python
# Get the image from which to extract Pk
image = (dat_img-best_model)/best_model
map_reso = (cluster.get_map_header()['CDELT2']*u.deg).to_value('radian')*cluster.D_ang.to_value('kpc')
```

```python
# Define a mask
mask = image*0 + 1
mask[best_model < np.amax(best_model)*0.05] = 0
mask_sm = gaussian_filter(mask, sigma=3)
plt.imshow(mask_sm, origin='lower')
```

```python
# Extract Pk
k3d, pk3d         = cluster.extract_p3d(image*mask_sm, map_reso, mask=mask_sm,
                                        method='Naive', statistic='mean', Nbin=200)
k3d_A12, pk3d_A12 = cluster.extract_p3d(image*mask, map_reso, mask=mask,
                                        method='Arevalo12', Nbin=200, unbias=True)
```

```python
# Try to match the model to the data by hand
cluster.model_pressure_fluctuation = {'name': 'CutoffPowerLaw',
                                      'Norm': 0.5, 'slope': -4.6,
                                      'Linj': 0.5*u.Mpc,'Ldis': 1*u.kpc}
kmod, pkmod = cluster.get_pressure_fluctuation_spectrum(kvec=np.logspace(-4,0, 100)*u.kpc**-1)
```

```python
# Show the Pk
plt.figure(figsize=(9,6))
plt.loglog(k3d.to_value('kpc-1'), np.sqrt(4*np.pi*k3d**3*pk3d).to_value(''))
plt.loglog(k3d_A12.to_value('kpc-1'), np.sqrt(4*np.pi*k3d_A12**3*pk3d_A12).to_value(''))
plt.loglog(kmod.to_value('kpc-1'), np.sqrt(4*np.pi*kmod**3*pkmod).to_value(''))
plt.xlabel(r'$k$ (kpc$^{-1}$)')
plt.ylabel(r'$\sqrt{4 \pi k^3 P(k)}$')
plt.ylim(5e-2,1e-0)
plt.xlim(1e-4,1e-1)
```
