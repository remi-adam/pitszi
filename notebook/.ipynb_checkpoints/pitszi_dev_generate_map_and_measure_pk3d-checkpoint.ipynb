```python
import sys
sys.path.insert(0,'/Users/adam/Project/NIKA/Software/Processing/Labtools/RA/pitszi/')
import numpy as np
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
import astropy.units as u
import numpy as np
from astropy.wcs import WCS
from scipy.ndimage import gaussian_filter

import pitszi
```

# Cluster model

```python
c = pitszi.Model(RA=0*u.deg, Dec=0*u.deg, redshift=0.4, M500=1e15*u.Msun, silent=True)

c.set_pressure_profile_universal_param('A10MD')

c.model_pressure_fluctuation = {'name': 'CutoffPowerLaw',
                                'Norm': 0.5, 
                                'slope': -11/3.0,
                                'Linj': 1*u.Mpc,
                                'Ldis': 1*u.kpc}

c.triaxiality = {'min_to_maj_axis_ratio':0.8, 'int_to_maj_axis_ratio':0.9,
                 'euler_angle1':40*u.deg, 'euler_angle2':50*u.deg, 'euler_angle3':45*u.deg}
        
c.map_coord  = SkyCoord(0*u.deg+0.2*u.arcmin, 0*u.deg, frame="icrs")
c.map_fov   = [5, 5]*u.arcmin
c.map_reso  = 5*u.arcsec
c.los_reso  = 20*u.kpc
c.los_size  = 5*u.Mpc
c.R_truncation = c.R500*10

seed = None

print('')
print('----- Scale -----')
print('Pixel size in kpc: ', (c.D_ang*(c.map_reso).to_value('rad')).to('kpc'))
print('FoV in Mpc: ', (c.D_ang*(c.map_fov[0]).to_value('rad')).to('Mpc'))
```

```python
r, p_r = c.get_pressure_profile(radius=np.logspace(1,4,100)*u.kpc)

k, pk = c.get_pressure_fluctuation_spectrum(kvec=np.logspace(-1,3,100)*u.Mpc**-1)

plt.figure(0, figsize=(18,6))
ax = plt.subplot(1, 2, 1)
plt.loglog(r, p_r)
plt.xlabel('r (kpc)')
plt.ylabel(r'$P_e(r))$ keV cm$^{-3}$')
plt.title('Mean pressure profile')

ax = plt.subplot(1, 2, 2)
plt.loglog(k.to_value('kpc-1'), np.sqrt(4*np.pi*k**3*pk).to_value(''))
plt.xlabel('k (1/kpc)')
plt.ylabel(r'$(4 \pi k^3 P(k))^{1/2}$')
plt.ylim(1e-2,5e-1)
plt.title('Fluctuation pressure power spectrum')
```

# Data realization

```python
fluct3d = c.pressure_cube_fluctuation(seed=seed)
prof3d  = c.pressure_cube_profile()

compton = c.get_sz_map(seed=seed)
compton_sph = c.get_sz_map(no_fluctuations=True)
```

```python
#----- Check the 3D Pk of data realization
Nx, Ny, Nz, proj_reso, proj_reso, los_reso = c.define_3dgrid()
k3d, pk3d = pitszi.utils.get_pk3d(fluct3d, proj_reso, los_reso, Nbin=50)
plt.loglog(k3d, (4*np.pi*k3d**3*pk3d)**0.5)
plt.ylim(1e-2,1e0)
```

```python
#----- Show the pressure fluctuation cube slice
plt.figure(figsize=(20,5))
ax = plt.subplot(1, 3, 1, projection=WCS(c.get_map_header()))
plt.imshow(fluct3d[:,:,int(fluct3d.shape[2]/2)].T, vmin=-1, vmax=1)
plt.colorbar()
plt.title('dP/P(r) at z=0')

ax = plt.subplot(1, 3, 2, projection=WCS(c.get_map_header()))
plt.imshow(prof3d[:,:,int(fluct3d.shape[2]/2)].T, vmin=0)
plt.colorbar()
plt.title('P(r) at z=0')

ax = plt.subplot(1, 3, 3, projection=WCS(c.get_map_header()))
plt.imshow((prof3d*(1+fluct3d))[:,:,int(fluct3d.shape[2]/2)].T)
plt.colorbar()
plt.title('P(r)+dP at z=0')
```

```python
plt.figure(figsize=(15,5))
ax = plt.subplot(1, 3, 1, projection=WCS(c.get_map_header()))
plt.imshow(compton.T*1e5)
plt.colorbar()
plt.title('True')

ax = plt.subplot(1, 3, 2, projection=WCS(c.get_map_header()))
plt.imshow(compton_sph.T*1e5)
plt.colorbar()
plt.title('Radial')

ax = plt.subplot(1, 3, 3, projection=WCS(c.get_map_header()))
plt.imshow(((compton-compton_sph)/compton_sph).T)
plt.colorbar()
plt.title('True fluctuation')
```

# Extract Pk

```python
image = (compton-compton_sph)/compton_sph
```

```python
kbin_edges = np.linspace(0, 2e-2, 31)
```

```python
k2d, pk2d  = pitszi.utils.get_pk2d(image, c.map_reso.to_value('radian')*c.D_ang.to_value('kpc'), 
                                       statistic='median', kedges=kbin_edges,
                                       Nbin=50)
k2dm, pk2dm  = pitszi.utils.get_pk2d(image, c.map_reso.to_value('radian')*c.D_ang.to_value('kpc'), 
                                       statistic='mean', kedges=kbin_edges,
                                       Nbin=50)
```

```python
conv = c.p3d_to_p2d_from_window_function()
```

```python
pk3d = pk2d*u.kpc**2 / np.mean(conv)
pk3dm = pk2dm*u.kpc**2 / np.mean(conv)
k3d = k2d*u.kpc**-1
```

```python
ktrue, pktrue = c.get_pressure_fluctuation_spectrum(kvec=np.logspace(-4,0, 100)*u.kpc**-1)
```

```python
plt.figure(figsize=(9,6))
plt.loglog(k3d.to_value('kpc-1'), np.sqrt(4*np.pi*k3d**3*pk3d).to_value(''))
plt.loglog(k3d.to_value('kpc-1'), np.sqrt(4*np.pi*k3d**3*pk3dm).to_value(''))
plt.loglog(ktrue.to_value('kpc-1'), np.sqrt(4*np.pi*ktrue**3*pktrue).to_value(''))
plt.xlabel(r'$k$ (kpc$^{-1}$)')
plt.ylabel(r'$\sqrt{4 \pi k^3 P(k)}$')
plt.ylim(1e-4,1e0)
plt.xlim(1e-4, 1e-1)
```

# Run MC realization

```python
Nmc = 30
Nbin = 50
pk2d_mc = np.zeros((Nmc,Nbin))
c.silent = True
for imc in range(Nmc):
    compton_i = c.get_sz_map()
    image_i = (compton_i-compton_sph)/compton_sph
    k2d_i, pk2d_i  = pitszi.utils.get_pk2d(image_i, 
                                           c.map_reso.to_value('radian')*c.D_ang.to_value('kpc'), 
                                           statistic='mean',Nbin=Nbin)
    pk2d_mc[imc, :] = pk2d_i
```

```python
pk3d_mc = pk2d_mc*u.kpc**2 / np.mean(conv)
pk3d_mean = np.mean(pk3d_mc, axis=0)
pk3d_std = np.std(pk3d_mc, axis=0)
```

```python
plt.figure(figsize=(9,6))
plt.loglog(k3d.to_value('kpc-1'), np.sqrt(4*np.pi*k3d**3*pk3d_mean).to_value(''), color='k', ls='-')
plt.loglog(k3d.to_value('kpc-1'), np.sqrt(4*np.pi*k3d**3*(pk3d_mean+pk3d_std)).to_value(''), color='k', ls='--')
plt.loglog(k3d.to_value('kpc-1'), np.sqrt(4*np.pi*k3d**3*(pk3d_mean-pk3d_std)).to_value(''), color='k', ls='--')
for imc in range(Nmc):
    plt.loglog(k3d.to_value('kpc-1'), np.sqrt(4*np.pi*k3d**3*(pk3d_mc[imc,:])).to_value(''), color='k', alpha=0.2)
    
plt.loglog(ktrue.to_value('kpc-1'), np.sqrt(4*np.pi*ktrue**3*pktrue).to_value(''), color='red')
plt.xlabel(r'$k$ (kpc$^{-1}$)')
plt.ylabel(r'$\sqrt{4 \pi k^3 P(k)}$')
plt.ylim(1e-4,1e0)
plt.xlim(0.5*np.amin(k3d.to_value('kpc-1')), 1.5*np.amax(k3d.to_value('kpc-1')))
```
