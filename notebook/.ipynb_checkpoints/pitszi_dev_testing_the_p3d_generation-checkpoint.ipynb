```python
def meas_pk(image, Nx, Ny, Nz, proj_reso, los_reso):

    fourier_image = np.fft.fftn(image)
    fourier_amplitudes = np.abs(fourier_image)**2

    k_x = np.fft.fftfreq(Nx, proj_reso)
    k_y = np.fft.fftfreq(Ny, proj_reso)
    k_z = np.fft.fftfreq(Nz, los_reso)
    k3d_x, k3d_y, k3d_z = np.meshgrid(k_x, k_y, k_z, indexing='xy')
    k3d_norm = np.sqrt(k3d_x**2 + k3d_y**2 + k3d_z**2)
    
    knrm = k3d_norm.flatten()
    fourier_amplitudes = fourier_amplitudes.flatten()

    kbins = np.linspace(np.amin([1/(proj_reso*Nx), 1/(proj_reso*Ny), 1/(los_reso*Nz)]),
                        np.amin(np.array([np.amax(k_x), np.amax(k_y), np.amax(k_z)])), 100)
    kvals = 0.5 * (kbins[1:] + kbins[:-1])
    Abins, _, _ = stats.binned_statistic(knrm, fourier_amplitudes,
                                         statistic = "mean",
                                         bins = kbins)
    #Abins *= np.pi * (kbins[1:]**2 - kbins[:-1]**2)

    return kvals, Abins/(Nx*Ny*Nz)
```

```python
def generate_isotropic_powerlaw_fluctuation_cube(size, alpha=2.0, rms=1.0, seed=None):
    """
    Generate a 3D fluctuation cube with isotropic power-law power spectrum.

    Parameters:
    - size: The size of the cube in each dimension (e.g., (128, 128, 128) for a 128x128x128 cube).
    - alpha: The power-law exponent for the power spectrum (default is 2.0 for Kolmogorov turbulence).
    - rms: The desired root mean square (RMS) of the fluctuations (default is 1.0).
    - seed: Seed for random number generation (optional).

    Returns:
    - A 3D NumPy array representing the fluctuation cube.
    """
    if seed is not None:
        np.random.seed(seed)

    # Create a frequency grid
    freq_x = np.fft.fftfreq(size[0])
    freq_y = np.fft.fftfreq(size[1])
    freq_z = np.fft.fftfreq(size[2])

    # Calculate the magnitude of the wave vector
    k = np.sqrt(freq_x[:, np.newaxis, np.newaxis]**2 +
                freq_y[np.newaxis, :, np.newaxis]**2 +
                freq_z[np.newaxis, np.newaxis, :]**2)

    # Calculate the power spectrum amplitude based on the power-law
    amplitude = k**-alpha
    amplitude[k==0] = 0

    # Normalize the power spectrum to achieve the desired RMS
    power_rms = np.sqrt(np.mean(amplitude**2))
    amplitude *= rms / power_rms
    
    # Generate random complex values with unit variance
    real_part = np.random.randn(*size)
    imag_part = np.random.randn(*size)

    # Apply the power spectrum to the Fourier coefficients
    real_part_fft = np.fft.fftn(real_part)
    imag_part_fft = np.fft.fftn(imag_part)
    real_part_fft *= amplitude
    imag_part_fft *= amplitude

    # Inverse Fourier transform to get the fluctuation cube
    fluctuation_cube = np.fft.ifftn(real_part_fft + 1j * imag_part_fft)

    return np.real(fluctuation_cube)
```

```python

size = (128, 128, 128)
desired_rms = 1.0  # Desired RMS value
alpha=2
fluctuation_cube = generate_isotropic_powerlaw_fluctuation_cube(size, alpha=alpha, rms=desired_rms)
```
